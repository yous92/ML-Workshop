# Unsupervised Learning

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

## 1. K-Means Clustering

```
from sklearn.cluster import KMeans
```

```
# Data Generation
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=100, centers=3, cluster_std=0.4, random_state=0)
plt.scatter(X[:, 0], X[:, 1])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Generated Data with 3 Clusters')
plt.show()
```

```
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Create and fit the KMeans model
model = KMeans(n_clusters=3, n_init=10)  # n_init: number of iterations for the algorithm;
model.fit(X)

# Predict cluster labels
labels = model.predict(X)

# Plot the data points colored by their cluster assignment
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', label='Data Points')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], c='r', marker='x', s=100, label='Centroids')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('KMeans Clustering')
plt.legend()
plt.show()

# Display the model inertia
print("Inertia:", model.inertia_)  # Sum of squared distances to nearest cluster center
```

### Elbow Method

```
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# List to store the inertia values for different k
inertia = []

# Range of k values
K_range = range(1, 20)

# Fit KMeans model for each k and store the inertia
for k in K_range:
    model = KMeans(n_clusters=k, n_init=10)  # n_init: number of iterations for the algorithm
    model.fit(X)
    inertia.append(model.inertia_)  # Calculate and store the inertia

# Plot inertia vs. number of clusters
plt.plot(K_range, inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Model Cost (Inertia)')
plt.title('Inertia vs. Number of Clusters')
plt.show()
```

The optimal number of clusters is 3.
